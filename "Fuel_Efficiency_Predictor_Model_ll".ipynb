# STEP#4: CREATE TESTING AND TRAINING DATASET

```python
y =  Fueleconomy['Fuel Economy (MPG)']
```

```python
X = Fueleconomy[['Horse Power']]
```

```python
X
```

```python
y
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Now you can split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
X_train.shape
```

```python
X_test.shape
```

# STEP#5: TRAIN THE MODEL

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression(fit_intercept = True)
```

```python
regressor.fit(X_train,y_train)
```

```python
print('Linear Model Coeff (m):', regressor.coef_)
print('Linear Model Coeff (b):', regressor.intercept_)
```

# STEP#6: TEST THE MODEL

```python
y_predict = regressor.predict(X_test)
y_predict
```

```python
y_test
```

```python
plt.scatter(X_train, y_train, color = 'gray')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.xlabel('Horse Power (HP)')
plt.ylabel('MPG')
plt.title('HP vs. MPG (Training Set)')
```

```python
HP = 500
# MPG?
# Reshape HP into a 2D array with one column
HP = np.array([[HP]])
# Now you can predict MPG
MPG = regressor.predict(HP)
MPG
```
